### GBM untuk Mitsubishi df1 dan NISSAN df2 

```python
#plot close price day
close_mit = df1['Close']
close_nis = df2['Close']

time_mit = np.linspace(1, len(close_mit), len(close_mit))
time_nis = np.linspace(1, len(close_nis), len(close_nis))

plt.plot(time_mit, close_mit, label = 'Mitsubishi')
plt.plot(time_nis, close_nis, label = 'Nissan')
plt.legend(loc = 'upper left')
```

```python
# PARAMETER PADA GBM
# GBM Exact Solution

# So:     initial stock price
# mu:     returns (drift coefficient)
# sigma:  volatility (diffusion coefficient)
# T:      time period
# N:      number of increments
# W:      brownian motion
```

```python
def daily_return(close):
    returns = []
    for i in range(0, len(close)-1):
        today = close[i+1]
        yesterday = close[i]
        daily_return = (today - yesterday)/yesterday
        returns.append(daily_return)
    return returns
```

```python
#Mitsubishi df1

returns_mit = daily_return(close_mit)

#anggap 1 tahun = 252 waktu untuk trading 
mu_mit = np.mean(returns_mit)*252.           # drift coefficient / returns
sig_mit = np.std(returns_mit)*np.sqrt(252.)  # diffusion coefficient / volatility

print("Mitsubishi mu     =", mu_mit)
print ("Mitsubishi sigma    =", sig_mit)

```

```python
#Nissan df2

returns_nis = daily_return(close_nis)

#trade disesuikan dengan jumlah data = 252 
mu_nis = np.mean(returns_nis)*252.           # drift coefficient / returns
sig_nis = np.std(returns_nis)*np.sqrt(252.)  # diffusion coefficient / volatility

print("Nissan mu     =", mu_nis)
print("Nissan sigma    =", sig_nis)
```

```python
seed = 5     
N  = 2.**6     # increments

def Brownian(seed, N):
    
    np.random.seed(seed)                         
    dt = 1./N                                         # time step
    b = np.random.normal(0., 1., int(N))*np.sqrt(dt)  # brownian  increments
    W = np.cumsum(b)                                  # brownian path
    return W, b


# brownian increments
b = Brownian(seed, N)[1]

# brownian motion
W = Brownian(seed, N)[0]
W = np.insert(W, 0, 0.)                      # W_0 = 0. untuk brownian motion
```

```python
def GBM(So, mu, sigma, W, T, N):    
    t = np.linspace(0.,1.,N+1)
    S = []
    S.append(So)
    for i in range(1,int(N+1)):
        drift = (mu - 0.5 * sigma**2) * t[i]
        diffusion = sigma * W[i-1]
        S_temp = So*np.exp(drift + diffusion)
        S.append(S_temp)
    return S, t
```

```python
# Jika seed = 22
seed = 22
T = 1
N = 2.**6
W = Brownian(seed, N)[0]

# Mitsubishi
So_mit = close_mit[0]  
soln_mit = GBM(So_mit, mu_mit, sig_mit, W, T, N)[0]     # Exact solution
t_mit = GBM(So_mit, mu_mit, sig_mit, W, T, N)[1]        # time increments for  plotting

#Nissan
So_nis = close_nis[0]  
soln_nis = GBM(So_nis, mu_nis, sig_nis, W, T, N)[0]     # Exact solution
t_nis = GBM(So_nis, mu_nis, sig_nis, W, T, N)[1]        # time increments for  plotting

plt.plot(t_mit, soln_mit, label = 'Mitsubishi')
plt.plot(t_nis, soln_nis, label = 'Nissan')
plt.ylabel('Stock Price Close, $')
plt.title('Geometric Brownian Motion')
plt.legend(loc = 'upper right')
```

```python
#Perbandingan actual nissan dengan seed=22
time = np.linspace(1, len(close_nis), len(close_mit))
time = [i/2422. for i in time]

plt.plot(time, close_nis, label = 'Actual')
plt.plot(t_nis, soln_nis, label = 'GBM (seed = 22)', ls='--')

plt.ylabel('Nissan Stock Price, $')
plt.title('Geometric Brownian Motion - Nissan')

plt.legend(loc = 'upper right')
```

#### Memperkirakan Best Seeds

```python
import statistics as st

actual=[0]*3

value_mit=[10000]*100
value_nis=[10000]*100

actual[1] = st.mean(close_mit)
actual[2] = st.mean(close_nis)

for i in range(1,99):
  seed = i
  W = Brownian(seed, N)[0]

  soln_mit = GBM(So_mit, mu_mit, sig_mit, W, T, N)[0]
  soln_nis = GBM(So_nis, mu_nis, sig_nis, W, T, N)[0]

  value_mit[i] = abs(st.mean(soln_mit) - actual[1])
  value_nis[i] = abs(st.mean(soln_nis) - actual[2])

best_mit=np.where(value_mit==min(value_mit))
best_mit=int(best_mit[0])
best_nis=np.where(value_nis==min(value_nis))
best_nis=int(best_nis[0])


print("Prediksi untuk Mitsubishi seed terbaik =", best_mit)
print("Prediksi untuk Nissan seed terbaik =",best_nis)
```

### Euler-Maruyama Model

```python
# Mitsubishi
seed = 5
So_mit = close_mit[0]           
W = Brownian(seed, N)[0]
T = 1.
N = 2.**6

soln_mit_em = GBM(So_mit, mu_mit, sig_mit, W, T, N)[0]    # Exact solution
t_mit = GBM(So_mit, mu_mit, sig_mit, W, T, N)[1]       # time increments for  plotting
plt.plot(t_mit, soln_mit_em, label ='GBM-Mit')


# Euler Maruyama Approximation
def EM(So, mu, sigma, b, T, N, M):
    dt = M * (1/N)  # EM step size
    L = N / M
    wi = [So]
    for i in range(0,int(L)):
        Winc = np.sum(b[(M*(i-1)+M):(M*i + M)])
        w_i_new = wi[i]+mu*wi[i]*dt+sigma*wi[i]*Winc
        wi.append(w_i_new)
    return wi, dt


# dt = 0.03125
b = Brownian(5, N)[1]    # Brownian increments 
M = 2                    
L = N/M
EM_approx_1 = EM(So_mit, mu_mit, sig_mit, b, T, N, M)[0]
time_EM_1 = np.linspace(0.,1.,L+1)

# dt = 0.0625
b = Brownian(5, N)[1]    # Brownian increments 
M = 4                    
L = N/M
EM_approx_2 = EM(So_mit, mu_mit, sig_mit, b, T, N, M)[0]
time_EM_2 = np.linspace(0.,1.,L+1)

plt.plot(time_EM_1, EM_approx_1, label = 'dt = 0.03125', ls ='--')
plt.plot(time_EM_2, EM_approx_2, label = 'dt = 0.0625', ls ='--')


plt.ylabel('Mitsubishi Stock Price, $')
plt.title('Geometric Brownian Motion - Mitsubishi')

plt.legend(loc = 'upper left')
```

```python
# Nissan
seed = 5
So_nis = close_nis[0]           
W = Brownian(seed, N)[0]
T = 1.
N = 2.**6

soln_nis_em = GBM(So_nis, mu_nis, sig_nis, W, T, N)[0]    # Exact solution
t_nis = GBM(So_nis, mu_nis, sig_nis, W, T, N)[1]       # time increments for  plotting
plt.plot(t_nis, soln_nis_em, label ='GBM-Nis')


# Euler Maruyama Approximation
def EM(So, mu, sigma, b, T, N, M):
    dt = M * (1/N)  # EM step size
    L = N / M
    wi = [So]
    for i in range(0,int(L)):
        Winc = np.sum(b[(M*(i-1)+M):(M*i + M)])
        w_i_new = wi[i]+mu*wi[i]*dt+sigma*wi[i]*Winc
        wi.append(w_i_new)
    return wi, dt


# dt = 0.03125
b = Brownian(5, N)[1]    # Brownian increments 
M = 2                    
L = N/M
EM_approx_1 = EM(So_nis, mu_nis, sig_nis, b, T, N, M)[0]
time_EM_1 = np.linspace(0.,1.,L+1)

# dt = 0.0625
b = Brownian(5, N)[1]    # Brownian increments 
M = 4                    
L = N/M
EM_approx_2 = EM(So_nis, mu_nis, sig_nis, b, T, N, M)[0]
time_EM_2 = np.linspace(0.,1.,L+1)

plt.plot(time_EM_1, EM_approx_1, label = 'dt = 0.03125', ls ='--')
plt.plot(time_EM_2, EM_approx_2, label = 'dt = 0.0625', ls ='--')


plt.ylabel('Nissan Stock Price, $')
plt.title('Geometric Brownian Motion - nissan')

plt.legend(loc = 'upper left')
```
